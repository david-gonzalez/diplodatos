```python
import pandas
import numpy
import seaborn
import seaborn as sns
import matplotlib.pyplot as plt
```

    https://www.cdc.gov/growthcharts/percentile_data_files.htm
2. Graficar las curvas correspondientes a cada percentil para niños y niñas

```python
ds_talla = pandas.read_csv('lenageinf.csv')
ds_peso = pandas.read_csv('wtageinf.csv')
ds_talla_varones = ds_talla[ds_talla.Sex == "1"]
ds_talla_mujeres = ds_talla[ds_talla.Sex == "2"]
ds_peso_varones = ds_peso[ds_peso.Sex == 1]
ds_peso_mujeres = ds_peso[ds_peso.Sex == 2]
```

```python
print("Percentil 3 de Peso Mujeres")
grid = sns.JointGrid("Agemos", "P3", data=ds_peso_mujeres, xlim=(0, 32), ylim=(0, 12), size=7)
grid.plot_joint(plt.scatter, color="r")
```

```python
print("Percentil 3 de Peso Varones")

grid = sns.JointGrid("Agemos", "P3", data=ds_peso_varones, xlim=(0, 32), ylim=(0, 12), size=7)
grid.plot_joint(plt.scatter, color="b")


```
