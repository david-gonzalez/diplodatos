# Obtenemos claves únicas:

De lo revisado, entendemos que la clave unica esta formada por: Sex, Agemo y el percentil. Transformaremos los datos segun esto:

```python
# Obtenemos solo las columnas que nos interesan: 
columns_to_ignore = ['L', 'M', 'S', 'Diff3', 'Diff5', 'Diff10', 'Diff25', 'Diff50', 'Diff75', 'Diff90', 'Diff95', 'Diff97']
columns_that_matter = [x for x in ds_talla.columns.tolist() if x not in columns_to_ignore ]
columns_that_matter
```

```python
# Transformamos los datos para obtener los valores de los percentiles segun la clave única:
ds_talla_percentiles = pandas.melt(ds_talla[columns_that_matter], id_vars=['Sex','Agemos'], var_name='Percentiles')
ds_talla_percentiles
```

Ahora el dataframe está listo para graficar:

## Acerca de la codificación del campo Sex

Sobre la codificación del campo Sex, obtuvimos de la pagina https://www.cdc.gov/growthcharts/percentile_data_files.htm que los valores son:

<H2>&#9658; 1 = male<br/>&#9658; 2 = female</H2>

Estas codificaciónes las utilizaremos luego para filtrar a la hora de mostrar los gráficos por cada sexo.

# Gráfico de Alturas por Edad en Niños

```python
plt.close()
fig, ax = plt.subplots(nrows=1, ncols=1)
ax = sns.pointplot(x="Agemos", y="value", hue='Percentiles', data=ds_talla_percentiles[ds_talla_percentiles['Sex']==1], markers=".")
ax.set_title("Boys Lenght")
ax.set_xlabel("Age");
ax.set_ylabel("Length");
ax.set_xticklabels(ax.get_xticklabels());
plt.show()
```

# Gráfico de Alturas por Edad en Niñas

```python
plt.close()
fig, ax = plt.subplots(nrows=1, ncols=1)
ax = sns.pointplot(x="Agemos", y="value", hue='Percentiles', data=ds_talla_percentiles[ds_talla_percentiles['Sex']==2], markers=".")
ax.set_title("Girls Lenght")
ax.set_xlabel("Age");
ax.set_ylabel("Length");
ax.set_xticklabels(ax.get_xticklabels());
plt.show()
```
