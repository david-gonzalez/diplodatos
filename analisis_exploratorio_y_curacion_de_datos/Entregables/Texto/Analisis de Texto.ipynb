```python
text = open('./millyunanoches.txt').read()

print(text.find("LAS MIL NOCHES Y UNA NOCHE\n\n\n_Tomo I"))
print(text.find("End of the Project Gutenberg"))
text = text[1169:369956]
raw_text = text
# text = text.lower()

punctuations = list(string.punctuation)
punctuations.append("_")
punctuations.append("--")
punctuations.append("«")
punctuations.append("»")
punctuations.append("á")
punctuations.append("_")
punctuations.append(".")
```

```python
tokens = nltk.word_tokenize(text)
```

```python
print("Total de palabras")
len(tokens)
```

```python
filtered_words = [word for word in tokens if word not in stopwords.words('spanish')]
print("Total de palabras sin stopwords")
len(filtered_words)
```

```python
filtered_words = [word for word in filtered_words if word not in punctuations]
print("Total de palabras sin caracteres")
len(filtered_words)
```

```python
print("Total de palabras sin contar las stopwords:")
len(filtered_words)
```

```python
dist = FreqDist(filtered_words)
len("Palabras únicas")
len(dist)
```

```python
print("Las palabras más frecuentes son: ")
print(dist.most_common())
```

```python
### Las funciones para detectar Nombre, están hechas para el idioma inglés. El siguiente ejemplo de una función
### Que busca nombres, con varios ejemplos en Stackoverflow tiene resultados muy malos para el español.

def get_human_names(tok):
    tokens = tok
    pos = nltk.pos_tag(tokens)
    sentt = nltk.ne_chunk(pos, binary = False)
    person_list = []
    person = []
    name = ""
    for subtree in sentt.subtrees(filter=lambda t: t.label() == 'PERSON'):
        for leaf in subtree.leaves():
            person.append(leaf[0])
        for part in person:
            name += part + ' '
        if name[:-1] not in person_list:
            person_list.append(name[:-1])
        name = ''
        person = []

    return (person_list)
```
